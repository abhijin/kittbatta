```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_edges_from([
    ('a','b'), ('a','c'), ('a','f'),
    ('b','c'), ('b','g'), ('b','j'),
    ('c','d'), ('c','e'), 
    ('d','e'), ('d','f'),
    ('e','g'), ('e','f'),
    ('f','h'),
    ('h','i'),
    ('i','j')
])
pos=nx.spring_layout(G)
#pos=nx.spectral_layout(G)

print(pos)
nx.draw(G,pos=pos,with_labels=True)
```
